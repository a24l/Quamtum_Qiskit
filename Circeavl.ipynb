```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ, execute, assemble
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator
from qiskit.visualization import plot_histogram

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

```python
qc_ha = QuantumCircuit(7,6)
# encode inputs in qubits 0 and 1
qc_ha.x(6) 
qc_ha.barrier()
# use cnots to write the XOR of the inputs on qubit 2
qc_ha.h(0)
qc_ha.h(1)
qc_ha.h(2)
qc_ha.h(3)
qc_ha.h(4)
qc_ha.h(5)
qc_ha.h(6)
qc_ha.x(2)
qc_ha.cx(3,6)
qc_ha.cx(0,6)
qc_ha.x(3)
qc_ha.cx(5,6)
qc_ha.x(0)
qc_ha.cx(3,6)

qc_ha.barrier()

qc_ha.h(0)
qc_ha.h(1)
qc_ha.h(2)
qc_ha.h(3)
qc_ha.h(4)
qc_ha.h(5)
qc_ha.h(6)

qc_ha.barrier()
# extract outputs
qc_ha.measure(0,0) # extract XOR value
qc_ha.measure(1,1) # extract AND value
qc_ha.measure(2,2) # extract XOR value
qc_ha.measure(3,3) # extract XOR value
qc_ha.measure(4,4) # extract AND value
qc_ha.measure(5,5) # extract XOR value

qc_ha.draw()
```

```python
sim = Aer.get_backend('qasm_simulator')
qobj = assemble(qc_ha)
counts = sim.run(qobj).result().get_counts()
plot_histogram(counts)
```
